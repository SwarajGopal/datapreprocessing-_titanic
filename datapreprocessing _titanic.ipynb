```python
import seaborn as sns
import numpy as np 
import pandas as pd
from sklearn.preprocessing import StandardScaler, LabelEncoder
```

```python
titanic = sns.load_dataset('titanic')
```

```python
titanic
```

```python
#Check for missing values
print(titanic.isnull().sum())
```

```python
#  Handle missing values 
titanic.dropna(subset=['age', 'embarked','deck'], inplace=True) # Fill missing values in 'deck' with a placeholder 
```

```python
# Convert categorical columns to numerical using LabelEncoder 
label_encoder = LabelEncoder() 
categorical_cols = ['sex', 'embarked', 'class', 'who', 'deck', 'embark_town', 'alive', 'alone'] 
for col in categorical_cols: 
    titanic[col] = label_encoder.fit_transform(titanic[col])
```

```python
#  Scale numerical features 
scaler = StandardScaler()
numerical_cols = ['age', 'fare'] 
titanic[numerical_cols] = scaler.fit_transform(titanic[numerical_cols])
```

```python
#  Feature engineering 
# Here, you can add or modify features as needed, such as creating a 'family_size' feature 
titanic['family_size'] = titanic['sibsp'] + titanic['parch'] + 1 
```
